###Handling Outliers.

```python
# Column outliers.
columns = ['views', 'comments', 'duration']

n = 1
plt.figure(figsize=(18,12))

for i in columns:
  plt.subplot(3,3,n)
  n=n+1
  sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightblue'})
  sns.boxplot(data[i])
  plt.title(i)
  plt.tight_layout()
```

```python
# Treatment of outliers.
columns = ['views', 'comments', 'duration']
for i in columns:
  iqr =  data[i].quantile(0.75)-data[i].quantile(0.25)
  data[i]= data[i].mask(data[i]>(data[i].quantile(0.75)+1.5*iqr), data[i].mean())
for i in columns:
  iqr =  data[i].quantile(0.75)-data[i].quantile(0.25)
  data[i]= data[i].mask(data[i]>(data[i].quantile(0.75)+1.5*iqr), data[i].mean())
  n = 1
plt.figure(figsize=(10,10))

for i in columns:
  plt.subplot(3,3,n)
  n=n+1
  sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightblue'})
  sns.boxplot(data[i])
  plt.title(i)
  plt.tight_layout()
```

##### What all outlier treatment techniques have you used and why did you use those techniques?

I used the quantile method to find the interquartile range. i found 75th percentile (q3) and 25th percentile (q1) and subtracted q1 from q3 to get the iqr and removed the rest by taking the average.

###Dealing with multicollinearity.

```python
#Multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```
